# **Model Selection & Evaluation**

```python
Y= dataset["selling_price"]
X= dataset.drop(columns=["selling_price","fuel encoded"])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#Models
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
Models ={"LR":LinearRegression(),
         "SVM": SVR(),
         "DT": DecisionTreeRegressor(),
         "RF": RandomForestRegressor(),
        "GB": GradientBoostingRegressor()}
```

```python
# roc curve for classes
fpr = []
tpr = []
thresh =[]
n_classes = 5

from sklearn.metrics import accuracy_score, mean_absolute_error,r2_score,mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
plt.figure(figsize=(16, 9))
for model, clf in Models.items():
    print(f"Train and evaluate {model}")
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)


    print(f"R2 score {model}: {r2_score(y_test, y_pred)}")
    print(f"Mean Absolute Error for {model}: {mean_absolute_error(y_test, y_pred)}\n")
    print(f"The root mean squared error = {np.sqrt(mean_squared_error(y_test, y_pred))}")

```

```python


from sklearn.metrics import accuracy_score, mean_absolute_error, r2_score, mean_squared_error
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
# Dictionaries to store the performance metrics for each model
r2_scores = {}
mae_scores = {}
rmse_scores = {}

plt.figure(figsize=(16, 9))

for model, clf in Models.items():
    print(f"Train and evaluate {model}")
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)

    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))

    r2_scores[model] = r2
    mae_scores[model] = mae
    rmse_scores[model] = rmse

    print(f"R2 score {model}: {r2}")
    print(f"Mean Absolute Error for {model}: {mae}\n")
    print(f"The root mean squared error = {rmse}")

# Plotting the performance metrics
metrics = ['R2 Score', 'Mean Absolute Error', 'Root Mean Squared Error']
models = list(Models.keys())

fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Plot R2 scores
axs[0].bar(models, [r2_scores[model] for model in models])
axs[0].set_title('R2 Score')
axs[0].set_xticklabels(models, rotation=45)
axs[0].set_ylabel('R2 Score')

# Plot MAE scores
axs[1].bar(models, [mae_scores[model] for model in models])
axs[1].set_title('Mean Absolute Error')
axs[1].set_xticklabels(models, rotation=45)
axs[1].set_ylabel('Mean Absolute Error')

# Plot RMSE scores
axs[2].bar(models, [rmse_scores[model] for model in models])
axs[2].set_title('Root Mean Squared Error')
axs[2].set_xticklabels(models, rotation=45)
axs[2].set_ylabel('Root Mean Squared Error')

plt.tight_layout()
plt.show()
```
